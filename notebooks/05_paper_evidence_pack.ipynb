# Project Journal 05: Paper Evidence Pack

This entry captures the point where the project stops being only a collection of experimental runs and becomes a reusable paper-facing evidence pack with a locked claim, tables, ablations, and a final comparison panel.


## Context

The earlier journal entries established three things:

- the hybrid model beats the plain ML baseline on synthetic OOD data
- the benchmark-constrained field protocol gives a more plausible lateral support story than raw field outputs
- the direct `2010` benchmark audit is much stronger than the earlier qualitative-only field story

This stage turns those pieces into one compact evidence pack that is actually usable for a short paper.


```python
from pathlib import Path
import json
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from IPython.display import display

ROOT = Path.cwd().resolve()
if not (ROOT / 'runs').exists():
    ROOT = ROOT.parent.resolve()

summary = json.loads((ROOT / 'runs' / 'paper_evidence' / 'results' / 'paper_evidence_summary.json').read_text())
synthetic_df = pd.read_csv(ROOT / 'runs' / 'paper_evidence' / 'results' / 'paper_synthetic_table.csv')
field_df = pd.read_csv(ROOT / 'runs' / 'paper_evidence' / 'results' / 'paper_field_table.csv')
direct_df = pd.read_csv(ROOT / 'runs' / 'paper_evidence' / 'results' / 'paper_direct_table.csv')
ablation_df = pd.read_csv(ROOT / 'runs' / 'paper_evidence' / 'results' / 'paper_ablation_table.csv')
ROOT
```

## What Changed This Stage

The evidence pack adds four tangible outputs:

- one locked paper claim
- one synthetic summary table
- one field and direct-benchmark comparison table
- one ablation table driven by the actual field protocol rather than hand-wavy commentary

It also saves a direct `2010` comparison panel and a short methods/results block that can be reused in writing.


```python
print('Claim')
print(summary['claim'])
print('\nSynthetic summary')
display(synthetic_df)
print('Field summary')
display(field_df)
print('Direct 2010 summary')
display(direct_df)
print('Ablations')
display(ablation_df)
```

## Current Results

The strongest evidence in this pack is now easy to say precisely:

- on synthetic OOD data, the hybrid model stays ahead of plain ML and far ahead of the classical baseline
- on the p07 field sequence, constrained support overlap with the `2010` benchmark rises from `2001` to `2006`
- on the direct `2010` p10 audit, the constrained hybrid clearly beats the cross-equalized baseline

The ablation table is also useful because it keeps the story honest. In this run, the benchmark reservoir constraint matters a lot, while shared thresholds and uncertainty gating do not move the p07 protocol much.


```python
panel_path = ROOT / 'runs' / 'paper_evidence' / 'results' / 'figures' / 'paper_direct_2010_panel.png'
image = mpimg.imread(panel_path)
plt.figure(figsize=(14, 5))
plt.imshow(image)
plt.title('Direct 2010 paper panel')
plt.axis('off')
plt.show()

print((ROOT / 'runs' / 'paper_evidence' / 'results' / 'methods_results_block.txt').read_text())
```

## Open Problems

The evidence pack still leaves a few important limits visible:

- the raw field model is still not field-ready without benchmark-constrained postprocessing
- the support benchmark is a structural envelope, not a dense voxel label
- the uncertainty-gating ablation is weak in this specific setup, so it should not be oversold as the main reason the protocol works

That means the paper claim is strong enough to submit, but it should stay centered on support alignment and selectivity rather than unconstrained field inversion.
